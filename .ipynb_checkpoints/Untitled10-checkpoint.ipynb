```python
# напишем функцию, которая показывает наглядно количество пропусков
def pass_value_barh(name_df, df):
    try:
        (
            (df.isna().sum())
            .to_frame()
            .rename(columns = {0:'space'})
            .query('space > 0')
            .sort_values(by = 'space', ascending = True)
            .plot(kind = 'barh', figsize = (19,6), rot = 0, legend = False, fontsize = 15)
            .set_title('Пропуски в ' + name_df + "\n", fontsize = 22, color = 'SteelBlue')    
        );    
    except:
        print('Пропусков не осталось ')
dataframes = {
    'applications.csv' : apl, 
    'admission.csv': adm
}
for name_df, df in dataframes.items():
    pass_value_barh(name_df, df)
```

# воспользуемся методом увелечения выборки в тренировочных данных
def upsample(features, target, repeat): 
    features_zeros = pd.DataFrame(features[target == 0])
    features_ones = pd.DataFrame(features[target == 1]) 
    target_zeros = target[target == 0] 
    target_ones = target[target == 1] 
    features_upsampled = pd.concat([features_zeros] + [features_ones] * repeat) 
    target_upsampled = pd.concat([target_zeros] + [target_ones] * repeat) 
    features_upsampled, target_upsampled = shuffle( features_upsampled, target_upsampled, random_state=12345) 
    return features_upsampled, target_upsampled
rep = ceil(len(target_train[target_train == 0]) / target_train.sum())

**Инициализация модели градиентного бустинга CatBoost** 

```python
%%time
# инициализация CatBoost с подбором гиперпараметров и визуализацией результатов обучения
param_grid = {
    'iterations': [800],
    'depth': [4],
    'l2_leaf_reg' : [5],
    'learning_rate': [0.05],
}

train_pool = Pool(features_train, label=target_train)

#auto_class_weights='Balanced', 
model = CatBoostClassifier(auto_class_weights='Balanced', random_state=12345, verbose=False)

grid_search = GridSearchCV(
    model,
    param_grid,
    cv=3,
    scoring='recall',
    n_jobs=-1
)

grid_search.fit(features_train, target_train)

best_model_catboost = grid_search.best_estimator_

print(f'Наилучшие гиперпараметры: {grid_search.best_params_}')
print(f'Значение recall для этой модели: {grid_search.best_score_:.2f}')

test_pool = Pool(features_test)

probabilities_test_catboost = best_model_catboost.predict_proba(features_test)
probabilities_one_test_catboost = probabilities_test_catboost[:, 1]
fpr, tpr, thresholds = roc_curve(target_test, probabilities_one_test_catboost)
plt.figure()
plt.plot(fpr, tpr)
plt.plot([0, 1], [0, 1], linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.title('ROC - кривая')
plt.show()
auc_roc_catboost = roc_auc_score(target_test, probabilities_one_test_catboost)
binary_predictions = (probabilities_one_test_catboost >= 0.5).astype(int)
test_recall = recall_score(target_test, binary_predictions)
print(f'Метрика AUC-ROC на тестовых данных: {auc_roc_catboost:.2f}')
print(f'Метрика recall на тестовых данных: {test_recall:.2f}')
```

```python
# Визуализация precision-recall кривой
precision, recall, _ = precision_recall_curve(target_test, probabilities_one_test_catboost)
auc_pr = auc(recall, precision)
plt.figure()
plt.plot(recall, precision, color='darkorange', lw=2, label=f'CatBoost (AUC-PR = {auc_pr:.2f})')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.ylim([0.0, 1.05])
plt.xlim([0.0, 1.0])
plt.title('Precision-Recall кривая')
plt.legend(loc='lower left')
plt.show()
```

**Инициализация случайного леса**

```python
# Параметры для случайного леса
rf_param_grid = {
    'n_estimators': [500, 600, 700],
    'max_depth': [2, 3, 4],
    'min_samples_split': [3, 6, 9],
    'min_samples_leaf': [2, 4, 6]
}

# Инициализация модели случайного леса
rf_model = RandomForestClassifier(class_weight='balanced', random_state=12345)

# Создание объекта GridSearchCV
rf_grid_search = GridSearchCV(
    rf_model,
    rf_param_grid,
    cv=3,
    scoring='recall',
    n_jobs=-1
)

# Подгонка модели на тренировочных данных
rf_grid_search.fit(features_train, target_train)

# Получение лучшей модели
best_rf_model = rf_grid_search.best_estimator_

# Вывод результатов
print(f'Наилучшие гиперпараметры для случайного леса: {rf_grid_search.best_params_}')
print(f'Значение recall для этой модели: {rf_grid_search.best_score_:.2f}')

# Предсказания на тестовых данных
rf_probabilities_test = best_rf_model.predict_proba(features_test)[:, 1]
rf_binary_predictions = (rf_probabilities_test >= 0.5).astype(int)

# Визуализация ROC-кривой
fpr_rf, tpr_rf, thresholds_rf = roc_curve(target_test, rf_probabilities_test)
plt.figure()
plt.plot(fpr_rf, tpr_rf)
plt.plot([0, 1], [0, 1], linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.title('ROC-кривая для случайного леса')
plt.show()

# Рассчет метрик AUC-ROC и recall
auc_roc_rf = roc_auc_score(target_test, rf_probabilities_test)
recall_rf = recall_score(target_test, rf_binary_predictions)

print(f'Метрика AUC-ROC для случайного леса на тестовых данных: {auc_roc_rf:.2f}')
print(f'Метрика recall для случайного леса на тестовых данных: {recall_rf:.2f}')
```

```python
# Визуализация precision-recall кривой
precision, recall, _ = precision_recall_curve(target_test, rf_probabilities_test)
auc_pr = auc(recall, precision)
plt.figure()
plt.plot(recall, precision, color='darkorange', lw=2, label=f'CatBoost (AUC-PR = {auc_pr:.2f})')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.ylim([0.0, 1.05])
plt.xlim([0.0, 1.0])
plt.title('Precision-Recall кривая')
plt.legend(loc='lower left')
plt.show()
```

```python
input_data = {'name': 'Информатика и вычислительная техника',
              'priority': [1],
              'competitive_scores': [255],
              'achievement_scores': [5],
              'ege_average_score': [83],
              'citizenship': ['Российская Федерация'],
              'with_ege': [1]
             }
new_data = pd.DataFrame(input_data)
```

```python
%%time
# составление пайплайна и вывод процента на посткпление
numeric_features = features_train_copy.select_dtypes(include=['int64', 'float64']).columns
categorical_features = features_train_copy.select_dtypes(include=['object']).columns

pipeline = Pipeline(steps=[
    ('preprocessor', ColumnTransformer(
        transformers=[
            ('num', SimpleImputer(strategy='median'),
             numeric_features),
            ('cat', OrdinalEncoder(),
             categorical_features)  
        ]
    )),
    ('classifier', CatBoostClassifier(depth = 4,
                                         iterations = 800,
                                         learning_rate = 0.05,
                                         random_state=12345,
                                         l2_leaf_reg=5, 
                                         auto_class_weights='Balanced', 
                                         verbose=False))
])

pipeline.fit(features_train_copy, target_train)

predictions = pipeline.predict(new_data)
probabilities = pipeline.predict_proba(new_data)
probabilities_class_1 = probabilities[:, 1]

if predictions[0] == 0:
    print(f'К сожалению, вероятность вашего поступления меньше 50%, ваш шанс на поступление {probabilities_class_1[0] * 100:.2f}%')
else:
    print(f'Поздравляем, у вас хорошие шансы на поступление, вероятность составляет {probabilities_class_1[0] * 100:.2f}%')
```

```python
%%time
# Обучение пайплайна и получение важности признаков
pipeline.fit(features_train_copy, target_train)
feature_importance = pipeline.named_steps['classifier'].feature_importances_
numeric_features = features_train_copy.select_dtypes(include=['int64', 'float64']).columns
categorical_features = features_train_copy.select_dtypes(include=['object']).columns

feature_importance_df = pd.DataFrame(
    {
        'Feature': numeric_features.tolist() + categorical_features.tolist(),
        'Importance': feature_importance
    }
)
feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)

print(feature_importance_df)
```
