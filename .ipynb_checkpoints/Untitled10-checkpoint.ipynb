```python
# напишем функцию, которая показывает наглядно количество пропусков
def pass_value_barh(name_df, df):
    try:
        (
            (df.isna().sum())
            .to_frame()
            .rename(columns = {0:'space'})
            .query('space > 0')
            .sort_values(by = 'space', ascending = True)
            .plot(kind = 'barh', figsize = (19,6), rot = 0, legend = False, fontsize = 15)
            .set_title('Пропуски в ' + name_df + "\n", fontsize = 22, color = 'SteelBlue')    
        );    
    except:
        print('Пропусков не осталось ')
dataframes = {
    'applications.csv' : apl, 
    'admission.csv': adm
}
for name_df, df in dataframes.items():
    pass_value_barh(name_df, df)
```

```python
# разделим наши данные для обучения и проверки модели 
features = df.drop('target', axis = 1)
target = df.target
features_train, features_test, target_train, target_test = train_test_split(
    features, target, test_size = 0.25, random_state = 12345
)
```

```python
# закодируем категориальные признаки 
features_train_dum = pd.get_dummies(features_train, columns = ['name', 'citizenship'])
features_test_dum = pd.get_dummies(features_test, columns = ['name', 'citizenship'])
# проверим, чтобы признаки были одинаковые
if features_train_dum.columns.to_list() == features_test_dum.columns.to_list():
    print('Признаки совпадают')
else:
    diff_features = set(features_train_dum.columns) - set(features_test_dum.columns)
    print('Признаки: \n')
    print(*diff_features, 'не совпадают')
```

```python
features_train.info()
```

```python
# Нормализуем данные в тренировочных и тестовых данных
scaler = MinMaxScaler()

features_to_normalize = ['priority', 'competitive_scores', 'achievement_scores', 'ege_average_score']

features_train_normalized = features_train_dum.copy()
features_train_normalized[features_to_normalize] = scaler.fit_transform(features_train_dum[features_to_normalize])

features_test_normalized = features_test_dum.copy()
features_test_normalized[features_to_normalize] = scaler.transform(features_test_dum[features_to_normalize])

features_train = features_train_normalized.copy()
features_test = features_test_normalized.copy()

del features_train_normalized
del features_test_normalized
```

```python
# посмотрим на баланс целевого пизнака в тренировочных данных
plt.figure(figsize = (10, 6))
target_train.plot.hist()
plt.xlabel('Поступил или нет')
plt.ylabel('Количество студентов')
plt.title('Баланс целевого признака')
plt.grid(False)
plt.show()
```

На графике видно, что существует сильный дизбаланс в данных. Для того, чтобы искоренить его можно воспользоваться методом увелечения выборки или же обучить модели со взвешиванием классов

```python
# воспользуемся методом увелечения выборки в тренировочных данных
def upsample(features, target, repeat): 
    features_zeros = pd.DataFrame(features[target == 0])
    features_ones = pd.DataFrame(features[target == 1]) 
    target_zeros = target[target == 0] 
    target_ones = target[target == 1] 
    features_upsampled = pd.concat([features_zeros] + [features_ones] * repeat) 
    target_upsampled = pd.concat([target_zeros] + [target_ones] * repeat) 
    features_upsampled, target_upsampled = shuffle( features_upsampled, target_upsampled, random_state=12345) 
    return features_upsampled, target_upsampled
rep = ceil(len(target_train[target_train == 0]) / target_train.sum())
features_train, target_train = upsample(features_train, target_train, rep)
# посмотрим на баланс целевого пизнака в тренировочных данных
plt.figure(figsize = (10, 6))
target_train.plot.hist()
plt.xlabel('Поступил или нет')
plt.ylabel('Количество студентов')
plt.title('Баланс целевого признака')
plt.grid(False)
plt.show()
```

Теперь решена проблема дизбаланса в данных

**Инициализация модели случайного леса для классификации**

```python
%%time
param_grid = {
    'n_estimators': range(10, 30),
    'max_depth': range(7, 13),
    'max_features': ['sqrt', 'log2', None],
    'max_leaf_nodes': range(12, 20)
}

model = RandomForestClassifier(random_state=12345)

grid_search = GridSearchCV(
    model,
    param_grid,
    cv=3,
    scoring='roc_auc',
    n_jobs=-1
)

grid_search.fit(features_train, target_train)

best_model_forest = grid_search.best_estimator_

print(f'Наилучшие гиперпараметры: {grid_search.best_params_}')
print(f'Значение roc_auc для этой модели: {grid_search.best_score_}')

probabilities_test = best_model_forest.predict_proba(features_test)
probabilities_one_test = probabilities_test[:, 1]
fpr, tpr, thresholds = roc_curve(target_test, probabilities_one_test)
plt.figure()
plt.plot(fpr, tpr)
plt.plot([0, 1], [0, 1], linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.title('ROC - кривая')
plt.show()
auc_roc = roc_auc_score(target_test, probabilities_one_test)
print(f'Метрика AUC-ROC на тестовых данных: {auc_roc:.2f}')
```

**Инициализация модели градиентного бустинга CatBoost** 

```python
%%time
param_grid = {
    'iterations': [100, 200, 300],
    'depth': [4, 6, 8, 10],
    'learning_rate': [0.01, 0.05, 0.1],
    'l2_leaf_reg': [1, 3, 5, 7, 9]
}

# Преобразование данных в формат CatBoost
train_pool = Pool(features_train, label=target_train)

model = CatBoostClassifier(random_state=12345, verbose=False)

grid_search = GridSearchCV(
    model,
    param_grid,
    cv=3,
    scoring='roc_auc',
    n_jobs=-1
)

grid_search.fit(features_train, target_train)

best_model_catboost = grid_search.best_estimator_

print(f'Наилучшие гиперпараметры: {grid_search.best_params_}')
print(f'Значение roc_auc для этой модели: {grid_search.best_score_}')

# Преобразование тестовых данных в формат CatBoost
test_pool = Pool(features_test)

probabilities_test_catboost = best_model_catboost.predict_proba(features_test)
probabilities_one_test_catboost = probabilities_test_catboost[:, 1]
fpr, tpr, thresholds = roc_curve(target_test, probabilities_one_test_catboost)
plt.figure()
plt.plot(fpr, tpr)
plt.plot([0, 1], [0, 1], linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.title('ROC - кривая')
plt.show()
auc_roc_catboost = roc_auc_score(target_test, probabilities_one_test_catboost)
print(f'Метрика AUC-ROC на тестовых данных: {auc_roc_catboost:.2f}')
```
